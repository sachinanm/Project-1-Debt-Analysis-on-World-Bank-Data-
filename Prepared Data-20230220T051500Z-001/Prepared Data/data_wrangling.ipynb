```python
country_metadata = pd.read_csv("IDS Datasets/IDS-DRSCountries_WLD_CountryMetadata.csv")
```

```python
country_metadata.sample(n=5)
```

```python
income_groups = country_metadata["Income Group"].unique().tolist()
```

```python
regions = country_metadata["Region"].unique().tolist()
```

```python
lst =[]
```

```python
for item in income_groups:
    lst.append(revised_data.index[revised_data.country_name == item].tolist())
```

```python
len(lst)
```

```python
for item in regions:
    lst.append(revised_data.index[revised_data.country_name == item].tolist())
```

```python
len(lst)
```

```python
new_lst = list(chain.from_iterable(lst))
```

```python
len(new_lst)
```

```python
revised_data.drop(new_lst, inplace = True)
```

```python
revised_data.drop(revised_data.index[revised_data.country_name == "Latin America & Caribbean (excluding high income)"].tolist(), inplace = True)
```

```python
revised_data.drop(revised_data.index[revised_data.country_name == "Middle income"].tolist(), inplace = True)
```
